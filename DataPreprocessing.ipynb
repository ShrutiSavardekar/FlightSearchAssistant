```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

Rawdata = pd.read_csv('data/raw-flight-data.csv')
print(Rawdata.shape)
Rawdata.head(10)
```

```python
print("Airports Dataset Null Value Check ")
print(Rawdata.shape)
Rawdata.isnull().sum()
```

```python
# removing null values from the given dataset
dataFlights = Rawdata.dropna()
# deleting duplicate rows
dataFlights = dataFlights.drop_duplicates()
print(dataFlights.shape)
dataFlights.isnull().sum()
```

```python
dataFl = pd.read_csv('data/flights.csv')
print(dataFl.shape)
dataFl.head(10)
```

```python
df_Fl = dataFl.copy()
df_Fl.head()

# Dropped the unwanted columns ('DepDelay', 'ArrDelay')
df_Fl.drop(['DepDelay', 'ArrDelay'], axis = 1, inplace = True)
print(df_Fl.info())
print(df_Fl.value_counts)

print("\n Flights Dataset Null Value Check \n")

print(df_Fl.isnull().sum())
```

```python
print(df_Fl['Carrier'].nunique())
print(df_Fl['Carrier'].value_counts())
plt.figure(figsize = (12, 8))
sns.heatmap(df_Fl.corr(), annot = True)
plt.title('Correlation matrix for flights dataset', fontsize = 16)
```
